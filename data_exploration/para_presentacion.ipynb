```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
df_1= pd.DataFrame(
    {
        "PROFILE_NAME": ["Perfil 1", "Perfil 2", "Perfil 3", "Perfil 4", "Perfil 5", "Perfil 6"],
        "CURRENT_PROPORTION_PCT": [17.5, 16.4, 20.9, 17.3, 14.7, 13.2]
    },
    index= [i for i in range(0, 6)]
)
```

```python
fig, ax= plt.subplots()
ax.bar(
    df_1["PROFILE_NAME"],
    df_1["CURRENT_PROPORTION_PCT"],
    label= df_1["PROFILE_NAME"],
    color= ['tab:red', 'tab:blue', 'tab:green', 'tab:orange', 'tab:purple', 'tab:brown']
)
ax.set_ylabel("Proporción actual (%)")
ax.set_xlabel("Perfil")
ax.set_title("Composición actual por perfil poblacional")
plt.show()
```

```python
df_2= pd.DataFrame(
    {
        "PROFILE_NAME": ["Perfil 1", "Perfil 2", "Perfil 3", "Perfil 4", "Perfil 5", "Perfil 6"],
        "MEAN_SCORE": [242.3, 253.4, 258.5, 269.2, 275.6, 307.4]
    },
    index= [i for i in range(0, 6)]
)
```

```python
fig, ax= plt.subplots()
ax.bar(
    df_2["PROFILE_NAME"],
    df_2["MEAN_SCORE"],
    label= df_2["PROFILE_NAME"],
    color= ['tab:red', 'tab:blue', 'tab:green', 'tab:orange', 'tab:purple', 'tab:brown']
)
ax.set_ylabel("Puntaje promedio")
ax.set_xlabel("Perfil")
ax.set_title("Puntaje promedio por perfil")
plt.show()
```
